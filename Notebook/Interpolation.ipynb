# Interpolation

## Try to practice with simple arrays

```python
x1 = [0,2,18,55]
y1 = [1234, 1231, 2000, 4000]
x2 = [1,7, 33, 44]
```

## Use "plotly" to plot

```python
import plotly as py
from plotly.graph_objs import Scatter, Layout
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
import numpy as np

y2 = [0]*len(x1)

for x in range(len(x2)):
    for ref in range(len(x1)):
        if x2[x] > x1[ref] and x2[x] < x1[ref+1]:
            y2[x] = (y1[ref+1] - y1[ref])/ (x1[ref+1]-x1[ref])* (x2[x]-x1[ref]) + y1[ref]


import plotly.graph_objs as go
trace = go.Scatter(x=x1, y=y1, name='Original data')
trace2 = go.Scatter(x=x2, y=y2, name='Interpolation')
iplot([trace, trace2], filename='test')
```

```python
import pandas as pd
filename1 = "20170907_test_FMR_S12_with sample_Field 0~-2000G_Freq 6G_avg 200"
filename2 = "20170907_test_FMR_S12_without sample_Field 0~-2000G_Freq 6G_avg 200"
data1 = pd.read_csv(filename1,sep="\t")
data2 = pd.read_csv(filename2,sep="\t")
```

```python
y = data1["Field (G)"].values
x = data1.index.values
x_fit = data2.index.values
y_fit = []
x_fit_new = []

for i in range(len(x_fit)):
    for ref in range(len(x)-1):
        if x_fit[i] == x[ref]:
            y_fit += [y[ref]]
            x_fit_new += [x[ref]]
        elif x_fit[i] > x[ref] and x_fit[i] < x[ref+1]:
            y_fit += [(y[ref+1] - y[ref])/ (x[ref+1]-x[ref])* (x_fit[i]-x[ref]) + y[ref]]
            x_fit_new += [x_fit[i]]
        elif x_fit[i] < x[ref] and x_fit[i] > x[ref+1]:
            y_fit += [(y[ref+1] - y[ref])/ (x[ref+1]-x[ref])* (x_fit[i]-x[ref]) + y[ref]]
            x_fit_new += [x_fit[i]]

trace = go.Scatter(x=x, y=y, name='Original data')
trace2 = go.Scatter(x=x_fit_new, y=y_fit, name='Interpolation')
iplot([trace, trace2], filename='test')
```

```python
y2 = data2["Field (G)"].values
x2 = data2.index.values
x_compare = []
y_compare = []
for i in range(len(x2)):
    for j in range(len(x_fit_new)):
        if x_fit_new[j] == x2[i]:
            x_compare += [x_fit_new[j]]
            y_compare += [y2[i]-y_fit[j]]
trace3 = go.Scatter(x=x_compare, y=y_compare, name='Sub.')
iplot([trace3,trace2], filename='Result')
```
